# 10. Masking

```python
a = np.arange(100).reshape(10,10)
a
```

```python
z = np.zeros(a.shape,like = a)
z
```

```python
o = np.ones(a.shape,like = a)
o
```

1. mask Even numbers

```python
np.where(a%2==0,o,z)
```

```python
np.where(a%2==0,o,a)
```

```python
np.where(a%2==0,a,z)
```

2. Diagonal masking

```python
np.where(a==a.diagonal(),o,z)
```

```python
np.where(a==a.diagonal(),a,z)
```

```python
np.where(a==a.diagonal(),o,a)
```

```python
np.where(a==a.diagonal(),z,a)
```
